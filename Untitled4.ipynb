```python
def fact(n):
  if(n == 0):
    return 1
  else:
    return n* fact(n-1)
```

```python
def nth_fibonacci(n):

    # base case
    if n <= 1:
        return n

    # sum of the two preceding
    # Fibonacci numbers
    return nth_fibonacci(n - 1) + nth_fibonacci(n - 2)

#Driver Code Starts
n = 3
result = nth_fibonacci(n)
print(result)

#Driver Code Ends
```
