```python
# new parameter  for default argument
def func(greetings, name = 'you'):
  return f"{greetings} {name}"
print(func('hey'))
```

```python
def student_info(*args,**kwargs):
  print(args)
  print(kwargs)
student_info('ravi','chandra',age = 20,school = 'high')
name =('ravi','chandra')
detail = {'age':20,'school' : 'high'}
# we wanted to specify using *,** while passing into the function
student_info(name,detail)
```
